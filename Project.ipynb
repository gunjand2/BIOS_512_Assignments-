```r
The dataset I have selected is the palmer penguins dataset from one of our earlier lectures. First, I downloaded the tidyverse package and the palmerpeguins package. Then to tabulate the dataset and review it, I used the code penguins |> head(). Following this, to see the comparison between penguins species and body mass for each, I plotted a line graph to visualise it. For the analysis and visualisation, I used the code 
p = ggplot(penguins, aes(y = species, x = body_mass_g)) +
      geom_line()
p 
This gave an output of a visual of a line graph representing the two. 
Next, I wanted to plot a graph between the bill depth and the flipper length. Additionally, the species were highlighted in different colors. The visual was a geom point visual. This was achieved using the following code: 
p = ggplot(penguins, aes(x = bill_depth_mm, y = flipper_length_mm, color = species)) +
      geom_point(na.rm = T) +
      scale_color_brewer(palette = 'Set2') +
      labs(x = 'Bill depth, mm', y = 'Flipper length, mm', color = 'Species')
p 
```
